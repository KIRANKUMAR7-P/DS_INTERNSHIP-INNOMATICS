```python
import re

def clean_subtitle(subtitle_text):
    # Remove timestamps
    subtitle_text = re.sub(r'\d{2}:\d{2}:\d{2},\d{3} --> \d{2}:\d{2}:\d{2},\d{3}\r\n', '', subtitle_text)
    
    # Remove non-printable characters and special characters
    cleaned_text = re.sub(r'[^\x20-\x7E]', '', subtitle_text)
    cleaned_text = re.sub(r'[^\w\s]', '', cleaned_text)

    # Remove HTML tags and URLs
    cleaned_text = re.sub(r'<[^>]+>', '', cleaned_text)
    cleaned_text = re.sub(r'http\S+', '', cleaned_text)

    # Remove line breaks and extra spaces
    cleaned_text = re.sub(r'\r?\n', ' ', cleaned_text)
    cleaned_text = re.sub(r'\s+', ' ', cleaned_text)

    # Convert text to lowercase
    cleaned_text = cleaned_text.lower()

    # Remove any remaining irrelevant content (words with 1 or 2 characters)
    cleaned_text = re.sub(r'\b\w{1,2}\b', '', cleaned_text)

    return cleaned_text.strip()

```

```python
df1['subtitle_content'] = df1['file_content'].apply(clean_subtitle)
```

```python
import re
import string
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('omw-1.4')
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))

def preprocess(raw_text):
    # Removing special characters and digits
    text = re.sub("[^a-zA-Z]", " ", raw_text)


    # change text to lower case
    text = text.lower()
    # Remove special characters and punctuation
    text = ''.join([char for char in text if char not in string.punctuation])

    # tokenize into words
    words = word_tokenize(text)

    # Remove stopwords and non-alphabetic words, and convert to lowercase
    tokens = [word.lower() for word in words if word.isalpha() and word.lower() not in stop_words]
    cleaned_text = ' '.join(tokens)

    # Lemmatize the text
    lemmatizer = WordNetLemmatizer()
    lemmatized_tokens = [lemmatizer.lemmatize(word) for word in tokens]
    normalized_text = ' '.join(lemmatized_tokens)


    return normalized_text
```

```python
from tqdm import tqdm
tqdm.pandas(desc="Processing Text")
df1['normalized_text'] = df1['subtitle_content'].progress_apply(preprocess)
```

```python
# Define the document chunker function
def document_chunker(document, chunk_size=1500, overlap_size=200):
    tokens = word_tokenize(document)  # Tokenize the document
    num_tokens = len(tokens)
    chunks = []

    # Iterate over the tokens and create chunks with specified overlap
    for start in range(0, num_tokens, chunk_size - overlap_size):
        end = min(start + chunk_size, num_tokens)
        chunk = tokens[start:end]
        chunks.append({'name': row['name'], 'chunk_index': len(chunks) + 1, 'chunk_text': ' '.join(chunk)})

    return chunks
all_chunks = []
for index, row in df.iterrows():
    document = row['normalized_text']
    document_chunks = document_chunker(document)
    all_chunks.extend(document_chunks)

# Create a new DataFrame from the list of chunks
chunks_df = pd.DataFrame(all_chunks)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
from joblib import dump, load
from tqdm import tqdm


# Initialize TfidfVectorizer with tqdm progress bar
tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(tqdm(chunk_df['chunk_text'], desc="TF-IDF Vectorization"))

# Save TF-IDF vectorizer and matrix for future use
dump(tfidf_vectorizer, 'tfidf_vectorizer.joblib')
dump(tfidf_matrix, 'tfidf_matrix.joblib')
```

```python
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity
from joblib import dump, load
from scipy.sparse import csr_matrix

# Load TF-IDF vectorizer and matrix
tfidf_vectorizer = load('tfidf_vectorizer.joblib')
tfidf_matrix = load('tfidf_matrix.joblib')


# Example query
query = 'watch online video'

# Transform the query using the loaded TF-IDF vectorizer
query_vector = tfidf_vectorizer.transform([query])



# Convert tfidf_matrix and query_vector to sparse matrices
# tfidf_sparse = csr_matrix(tfidf_matrix)
# query_sparse = csr_matrix(query_vector)

# Calculate cosine similarity between the query vector and documents
cosine_similarities = cosine_similarity(query_vector, tfidf_matrix, dense_output=False).toarray().flatten()



# Get the indices of the top 5 most similar documents
top_indices = cosine_similarities.argsort()[-5:][::-1]

# Get the top 5 most similar documents and their cosine similarity scores
top_documents = []
for idx in top_indices:
    top_documents.append((chunk_df.iloc[idx], cosine_similarities[idx]))

# Initialize a list to store the required information
result_info = []

# Extract the required information from the top_documents list
for document, score in top_documents:
    result_info.append({
        'Cosine Similarity Score': score,
        'Name': document['name'],
        'Chunk': document['chunk_text'],
        'Chunk ID': document['chunk_index']
    })

# Print the required information
print("Top 5 most similar documents:")
for info in result_info:
    print("Cosine Similarity Score:", info['Cosine Similarity Score'])
    print("Name:", info['Name'])
    print("Chunk:", info['Chunk'])
    print("Chunk ID:", info['Chunk ID'])
    print("#"*50)


```
